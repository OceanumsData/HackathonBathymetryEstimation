```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
DATA_DIR = 'data_sample/'
SPOT = 'saint_louis/'
```

```python
filename = np.random.choice(os.listdir(DATA_DIR + SPOT + SPOT))
if 'RAW' in filename:
    raw_filename = filename
    processed_filename = filename[0:-8] + '.npy'
else:
    raw_filename = filename[0:-4] + '_RAW.npy'
    processed_filename = filename
```

```python
raw_img = np.load(DATA_DIR + SPOT + SPOT + raw_filename)
processed_img = np.load(DATA_DIR + SPOT + SPOT + processed_filename)
```

```python
rgb_raw_img = np.moveaxis(np.array([raw_img[:,:,3], raw_img[:,:,2], raw_img[:,:,0]]), 0, -1)
rgb_processed_img = np.moveaxis(np.array([processed_img[:,:,3], processed_img[:,:,2], processed_img[:,:,0]]), 0, -1)
```

```python
fig, ax = plt.subplots(2, 5, figsize=(25,10))
for i in range(4):
    ax[0,i].imshow(raw_img[:,:,i])
    ax[1,i].imshow(processed_img[:,:,i])
ax[0,4].imshow(rgb_raw_img)
ax[1,4].imshow(rgb_processed_img)
```

```python
plt.imshow((processed_img[:,:,3] - np.mean(processed_img[:,:,3]))/np.linalg.norm(processed_img[:,:,3]) - (processed_img[:,:,0] - np.mean(processed_img[:,:,0]))/np.linalg.norm(processed_img[:,:,0]))
```
